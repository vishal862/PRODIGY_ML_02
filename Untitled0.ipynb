```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df
```

```python
# --->df.iloc[row_start:row_end,column_start:column_end]
# --->if you want all the rows and some coloums then use this --> : in place of rows and specify coloums
# --->if you want all the coloumns and some rows then use this --> : in place of coloumns and specify rows

issdata=df.iloc[:,3:5].values
issdata
```

```python
# for finding the optimal value of K use a loop

from sklearn.cluster import KMeans
#ssbc --> sum of squares of distance between centroids
ssdbc=[]

for i in range(1,11):
  kmeans = KMeans(n_clusters = i,init='k-means++', random_state=0)
  kmeans.fit(issdata)
  ssdbc.append(kmeans.inertia_)
ssdbc
```

```python
#for plotting the elbow graph
plt.plot(range(1,11),ssdbc)
plt.title('Elboe Method')
plt.xlabel('K-Values')
plt.ylabel('ssdbc')
plt.legend()
plt.show()
```

```python
#build a model
kmeansmodel = KMeans(n_clusters = 5,init='k-means++', random_state=0)
```

```python
#start predicting
y_kmeans = kmeansmodel.fit_predict(issdata)
plt.scatter(issdata[y_kmeans==0,0],issdata[y_kmeans==0,1], s=80, c='green',label='Customer1')
plt.scatter(issdata[y_kmeans==1,0],issdata[y_kmeans==1,1], s=80, c='red',label='Customer2')
plt.scatter(issdata[y_kmeans==2,0],issdata[y_kmeans==2,1], s=80, c='blue',label='Customer3')
plt.scatter(issdata[y_kmeans==3,0],issdata[y_kmeans==3,1], s=80, c='yellow',label='Customer4')
plt.scatter(issdata[y_kmeans==4,0],issdata[y_kmeans==4,1], s=80, c='pink',label='Customer5')
plt.title('Clusturs of customers')
plt.xlabel('Annual Income ($k)')
plt.ylabel('Spending Scores(1-100)')
plt.legend()
plt.show()
```
